```python
import os
import pyspark
import warnings
warnings.filterwarnings("ignore")
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.master("local").appName("greenhub-farmer-spark-sesssion").getOrCreate()
```

```python
path = "raw_data\devices"
```

```python
sample_df = spark.read.parquet([f"{path}\{i}" for i in os.listdir(path)][0])
```

```python
df = sample_df.limit(0)
```

```python
df = df.limit(0)
for i in os.listdir(path):
    temp_df = spark.read.parquet(f"{path}\{i}")
    df = df.union(temp_df)
```

```python
print(f"Sample dataframe records count if {sample_df.count()}, All combined data frame count is {df.count()}")
```
